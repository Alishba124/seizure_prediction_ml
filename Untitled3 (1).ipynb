```python
# CONTEXT: Loading all 3 datasets
# dataset1 folder = UCI Epileptic, dataset2 folder = EEG Eye State, dataset3 folder = CHB-MIT Kaggle
import pandas as pd
import numpy as np
import os

os.makedirs('/content/plots', exist_ok=True)

# Dataset 1 - UCI Epileptic (in dataset2 folder)
df1 = pd.read_csv('/content/dataset2/Epileptic Seizure Recognition.csv')
df1['label'] = (df1['y'] == 1).astype(int)
df1.drop(columns=['y', 'Unnamed: 0'], inplace=True, errors='ignore')
df1 = df1.select_dtypes(include=[np.number])
print('df1:', df1.shape, '| Seizure rate:', df1['label'].mean().round(3))

# Dataset 2 - EEG Eye State (in dataset3 folder)
df2 = pd.read_csv('/content/dataset3/EEG_Eye_State_Classification.csv')
df2['label'] = df2['eyeDetection'].astype(int)
df2.drop(columns=['eyeDetection'], inplace=True)
df2 = df2.select_dtypes(include=[np.number])
print('df2:', df2.shape, '| Class rate:', df2['label'].mean().round(3))

# Dataset 3 - CHB-MIT from Kaggle (in dataset3 folder)
print('Loading CHB-MIT... please wait 2-3 mins')
df3 = pd.read_csv('/content/dataset3/EEG_Scaled_data.csv')
print('df3:', df3.shape)
print('✅ All 3 datasets loaded!')
```

```python
# Dataset 3 has 36,864 channels — way too many, sample 178 columns to match report
# Take a sample of 9500 rows and 178 random feature columns + target
df3_cols = [c for c in df3.columns if c != 'target']
df3_sample_cols = np.random.choice(df3_cols, size=178, replace=False)

df3 = df3[list(df3_sample_cols) + ['target']].sample(n=9500, random_state=42).reset_index(drop=True)

# Make seizure rate ~7% (currently unknown — check it)
print('df3 seizure rate:', df3['target'].mean().round(3))
print('df3 shape:', df3.shape)
```

```python
# Dataset 1
X1 = df1.drop(columns=['label']).values
y1 = df1['label'].values

# Dataset 2
X2 = df2.drop(columns=['label']).values
y2 = df2['label'].values

# Dataset 3
X3 = df3.drop(columns=['target']).values
y3 = df3['target'].values

print('X1:', X1.shape, '| y1 seizure rate:', y1.mean().round(3))
print('X2:', X2.shape, '| y2 class rate:', y2.mean().round(3))
print('X3:', X3.shape, '| y3 seizure rate:', y3.mean().round(3))
```

```python
from sklearn.model_selection import train_test_split

X1_tr, X1_val, y1_tr, y1_val = train_test_split(X1, y1, test_size=0.2, stratify=y1, random_state=42)
X2_tr, X2_val, y2_tr, y2_val = train_test_split(X2, y2, test_size=0.2, stratify=y2, random_state=42)
X3_tr, X3_val, y3_tr, y3_val = train_test_split(X3, y3, test_size=0.2, stratify=y3, random_state=42)

print('Splits done')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

# Pipeline A: Scale → SelectKBest → LogReg
pipe_A = Pipeline([
    ('scaler', StandardScaler()),
    ('select', SelectKBest(f_classif, k=30)),
    ('lr', LogisticRegression(C=1.0, max_iter=1000, random_state=42))
])

# Pipeline B: Scale → PCA → LogReg
pipe_B = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=30)),
    ('lr', LogisticRegression(C=1.0, max_iter=1000, random_state=42))
])

print('Pipelines defined')
```

```python
from sklearn.metrics import accuracy_score, f1_score, average_precision_score, roc_auc_score

datasets = {
    'UCI Epileptic':  (X1_tr, X1_val, y1_tr, y1_val),
    'EEG Eye State':  (X2_tr, X2_val, y2_tr, y2_val),
    'CHB-MIT Kaggle': (X3_tr, X3_val, y3_tr, y3_val),
}

print(f"{'Dataset':<20} {'Acc':>6} {'F1':>6} {'PR-AUC':>8} {'ROC-AUC':>9}")
print("-"*55)

for name, (Xtr, Xval, ytr, yval) in datasets.items():
    pipe_A.fit(Xtr, ytr)
    ypred = pipe_A.predict(Xval)
    yprob = pipe_A.predict_proba(Xval)[:,1]
    acc  = accuracy_score(yval, ypred)
    f1   = f1_score(yval, ypred)
    prauc= average_precision_score(yval, yprob)
    roc  = roc_auc_score(yval, yprob)
    print(f"{name:<20} {acc:>6.3f} {f1:>6.3f} {prauc:>8.3f} {roc:>9.3f}")
```

```python
from sklearn.metrics import accuracy_score, f1_score, average_precision_score, roc_auc_score

datasets = {
    'UCI Epileptic':  (X1_tr,X1_val,y1_tr,y1_val),
    'EEG Eye State':  (X2_tr,X2_val,y2_tr,y2_val),
    'CHB-MIT Kaggle': (X3_tr,X3_val,y3_tr,y3_val),
}

results = {}
print(f"{'Dataset':<20} {'Acc':>6} {'F1':>6} {'PR-AUC':>8} {'ROC-AUC':>9}")
print("-"*55)
for name,(Xtr,Xval,ytr,yval) in datasets.items():
    pipe_A.fit(Xtr,ytr)
    yp = pipe_A.predict(Xval)
    yprob = pipe_A.predict_proba(Xval)[:,1]
    acc  = accuracy_score(yval,yp)
    f1   = f1_score(yval,yp,zero_division=0)
    prauc= average_precision_score(yval,yprob)
    roc  = roc_auc_score(yval,yprob)
    results[name] = dict(Accuracy=acc,F1=f1,PR_AUC=prauc,ROC_AUC=roc)
    print(f"{name:<20} {acc:>6.3f} {f1:>6.3f} {prauc:>8.3f} {roc:>9.3f}")

# ── Bar chart ──────────────────────────────────────────────────────────────
import matplotlib.pyplot as plt
import numpy as np

names  = list(results.keys())
f1s    = [results[n]['F1']     for n in names]
praucs = [results[n]['PR_AUC'] for n in names]
accs   = [results[n]['Accuracy'] for n in names]

x = np.arange(len(names)); w = 0.25
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(x-w,  accs,  w, label='Accuracy', color='#0296A9')
ax.bar(x,    f1s,   w, label='F1-Score', color='#16A34A')
ax.bar(x+w,  praucs,w, label='PR-AUC',   color='#F96167')
ax.set_xticks(x); ax.set_xticklabels(names, fontsize=12)
ax.set_ylim(0,1.1); ax.set_ylabel('Score', fontsize=12)
ax.set_title('Baseline Model Performance Across Datasets', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
for i,vals in enumerate([accs,f1s,praucs]):
    for j,v in enumerate(vals):
        ax.text(j + (i-1)*w, v+0.01, f'{v:.2f}', ha='center', fontsize=9, fontweight='bold')
ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('/content/plots/baseline_results.png', dpi=150)
plt.show()
```

```python
from sklearn.model_selection import learning_curve

C_vals     = [0.0001,  1.0,   1000]
C_labels   = ['C=0.0001\n(Underfit)', 'C=1.0\n(Baseline)', 'C=1000\n(Overfit)']
colors     = ['#F5C542', '#16A34A', '#F96167']

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Learning Curves — Bias vs Variance Tradeoff (UCI Epileptic)',
             fontsize=14, fontweight='bold')

scaler = StandardScaler()
X1s = scaler.fit_transform(X1)

summary = []
for i, (C, label, col) in enumerate(zip(C_vals, C_labels, colors)):
    model = LogisticRegression(C=C, max_iter=1000, random_state=42)
    train_sizes, train_scores, val_scores = learning_curve(
        model, X1s, y1, cv=5, scoring='f1',
        train_sizes=np.linspace(0.1,1.0,10), n_jobs=-1)

    tr_mean = train_scores.mean(axis=1)
    val_mean = val_scores.mean(axis=1)
    tr_std   = train_scores.std(axis=1)
    val_std  = val_scores.std(axis=1)

    ax = axes[i]
    ax.plot(train_sizes, tr_mean,  color=col, lw=2.5, label='Train F1')
    ax.plot(train_sizes, val_mean, color=col, lw=2, ls='--', label='Val F1')
    ax.fill_between(train_sizes, tr_mean-tr_std,  tr_mean+tr_std,  alpha=0.1, color=col)
    ax.fill_between(train_sizes, val_mean-val_std, val_mean+val_std, alpha=0.1, color=col)
    ax.set_title(label, fontsize=13, fontweight='bold')
    ax.set_xlabel('Training set size', fontsize=11)
    ax.set_ylabel('F1-Score', fontsize=11)
    ax.set_ylim(0.1, 1.05)
    ax.legend(fontsize=10)
    ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)

    summary.append((label.replace('\n',' '), tr_mean[-1], val_mean[-1]))

plt.tight_layout()
plt.savefig('/content/plots/learning_curves.png', dpi=150)
plt.show()

print(f"\n{'Scenario':<25} {'Train F1':>9} {'Val F1':>8}")
print('-'*45)
for s,tr,v in summary:
    print(f"{s:<25} {tr:>9.3f} {v:>8.3f}")
```

```python
from sklearn.linear_model import SGDClassifier

reg_results = {}
dataset_items = {
    'UCI Epileptic':  (X1_tr,X1_val,y1_tr,y1_val),
    'EEG Eye State':  (X2_tr,X2_val,y2_tr,y2_val),
    'CHB-MIT Kaggle': (X3_tr,X3_val,y3_tr,y3_val),
}

for reg_name, penalty, l1r in [('L1','l1',1.0),('L2','l2',0.0),('Elastic Net','elasticnet',0.5)]:
    reg_results[reg_name] = {}
    for ds_name,(Xtr,Xval,ytr,yval) in dataset_items.items():
        sc = StandardScaler()
        Xtr_s  = sc.fit_transform(Xtr)
        Xval_s = sc.transform(Xval)

        clf = SGDClassifier(loss='log_loss', penalty=penalty, l1_ratio=l1r,
                            alpha=0.01, max_iter=1000, random_state=42)
        clf.fit(Xtr_s, ytr)
        yp    = clf.predict(Xval_s)
        yprob = clf.predict_proba(Xval_s)[:,1]

        f1    = f1_score(yval, yp, zero_division=0)
        prauc = average_precision_score(yval, yprob)
        sparsity = (clf.coef_==0).mean()
        reg_results[reg_name][ds_name] = dict(F1=f1, PR_AUC=prauc, Sparsity=sparsity)

# ── Print table ────────────────────────────────────────────────────────────
print(f"\n{'Regularizer':<14} {'Dataset':<20} {'F1':>6} {'PR-AUC':>8} {'Sparsity':>10}")
print('-'*62)
for reg in reg_results:
    for ds in reg_results[reg]:
        r = reg_results[reg][ds]
        print(f"{reg:<14} {ds:<20} {r['F1']:>6.3f} {r['PR_AUC']:>8.3f} {r['Sparsity']:>9.1%}")

# ── Chart: F1 grouped by regularizer ──────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Regularization Study: L1 vs L2 vs Elastic Net', fontsize=14, fontweight='bold')

metrics_plot = ['F1', 'PR_AUC', 'Sparsity']
titles = ['F1-Score', 'PR-AUC', 'Sparsity (% zero coefficients)']
colors_r = ['#F96167', '#0296A9', '#16A34A']
regs = list(reg_results.keys())
ds_names = list(dataset_items.keys())
x = np.arange(len(regs)); w = 0.25

for ax, metric, title in zip(axes, metrics_plot, titles):
    for i, ds in enumerate(ds_names):
        vals = [reg_results[r][ds][metric] for r in regs]
        ax.bar(x + (i-1)*w, vals, w, label=ds, color=colors_r[i])
    ax.set_xticks(x); ax.set_xticklabels(regs, fontsize=11)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.legend(fontsize=9); ax.set_ylim(0, 1.1)
    ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('/content/plots/regularization_study.png', dpi=150)
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, precision_recall_curve

sc3 = StandardScaler()
X3_tr_s  = sc3.fit_transform(X3_tr)
X3_val_s = sc3.transform(X3_val)


from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import precision_recall_curve

sc3 = StandardScaler()
X3_tr_s  = sc3.fit_transform(X3_tr)
X3_val_s = sc3.transform(X3_val)

strategies = {
    'No Resampling': (X3_tr_s, y3_tr),
    'SMOTE':         SMOTE(random_state=42).fit_resample(X3_tr_s, y3_tr),
    'Undersampling': RandomUnderSampler(random_state=42).fit_resample(X3_tr_s, y3_tr),
    'Class Weight':  (X3_tr_s, y3_tr),   # handled via model param
}

imb_results = {}
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Class Imbalance Strategies — Precision-Recall Curves (CHB-MIT)',
             fontsize=14, fontweight='bold')
axes = axes.flatten()
colors_i = ['#F96167','#0296A9','#F5C542','#16A34A']

for ax, (name, data), col in zip(axes, strategies.items(), colors_i):
    Xtr_i, ytr_i = data
    cw = 'balanced' if name == 'Class Weight' else None
    clf = LogisticRegression(C=1.0, max_iter=1000, random_state=42, class_weight=cw)
    clf.fit(Xtr_i, ytr_i)
    yp    = clf.predict(X3_val_s)
    yprob = clf.predict_proba(X3_val_s)[:,1]

    prec_c, rec_c, _ = precision_recall_curve(y3_val, yprob)
    prec = precision_score(y3_val, yp, zero_division=0)
    rec  = recall_score(y3_val, yp)
    f1   = f1_score(y3_val, yp, zero_division=0)
    prauc= average_precision_score(y3_val, yprob)
    imb_results[name] = dict(Precision=prec, Recall=rec, F1=f1, PR_AUC=prauc)

    ax.plot(rec_c, prec_c, color=col, lw=2.5)
    ax.fill_between(rec_c, prec_c, alpha=0.1, color=col)
    ax.set_title(f'{name}\nPrec={prec:.3f} | Rec={rec:.3f} | F1={f1:.3f}',
                 fontsize=11, fontweight='bold')
    ax.set_xlabel('Recall'); ax.set_ylabel('Precision')
    ax.set_xlim(0,1); ax.set_ylim(0,1.05)
    ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('/content/plots/imbalance_pr_curves.png', dpi=150)
plt.show()

# Summary table
print(f"\n{'Strategy':<18} {'Precision':>10} {'Recall':>8} {'F1':>6} {'PR-AUC':>8}")
print('-'*55)
for name,r in imb_results.items():
    print(f"{name:<18} {r['Precision']:>10.3f} {r['Recall']:>8.3f} {r['F1']:>6.3f} {r['PR_AUC']:>8.3f}")
```

```python
pipe_configs = {
    'A: Scale→SelectK→LR': Pipeline([('sc',StandardScaler()),('sel',SelectKBest(f_classif,k=30)),
                                      ('lr',LogisticRegression(C=1,max_iter=1000,random_state=42))]),
    'A-Rev: SelectK→Scale→LR': Pipeline([('sel',SelectKBest(f_classif,k=30)),('sc',StandardScaler()),
                                          ('lr',LogisticRegression(C=1,max_iter=1000,random_state=42))]),
    'B: Scale→PCA→LR':    Pipeline([('sc',StandardScaler()),('pca',PCA(n_components=30)),
                                    ('lr',LogisticRegression(C=1,max_iter=1000,random_state=42))]),
    'B-Rev: PCA→Scale→LR':Pipeline([('pca',PCA(n_components=30)),('sc',StandardScaler()),
                                     ('lr',LogisticRegression(C=1,max_iter=1000,random_state=42))]),
}

pipe_res = {}
print(f"{'Pipeline':<28} {'F1':>6} {'PR-AUC':>8} {'Accuracy':>10}")
print('-'*56)
for pname, pipe in pipe_configs.items():
    try:
        pipe.fit(X1_tr, y1_tr)
        yp    = pipe.predict(X1_val)
        yprob = pipe.predict_proba(X1_val)[:,1]
        f1    = f1_score(y1_val,yp,zero_division=0)
        prauc = average_precision_score(y1_val,yprob)
        acc   = accuracy_score(y1_val,yp)
        pipe_res[pname] = dict(F1=f1,PR_AUC=prauc,Accuracy=acc)
        print(f"{pname:<28} {f1:>6.3f} {prauc:>8.3f} {acc:>10.3f}")
    except Exception as e:
        print(f"{pname:<28} ERROR: {e}")

# ── Bar chart ──────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(11,5))
pnames = list(pipe_res.keys())
f1_p   = [pipe_res[p]['F1']    for p in pnames]
pr_p   = [pipe_res[p]['PR_AUC'] for p in pnames]
x = np.arange(len(pnames)); w = 0.3
colors_p = ['#16A34A','#F96167','#16A34A','#F96167']
ax.bar(x-w/2, f1_p,  w, color=colors_p, label='F1', alpha=0.9)
ax.bar(x+w/2, pr_p,  w, color=colors_p, label='PR-AUC', alpha=0.5, hatch='//')
for i,(f,p) in enumerate(zip(f1_p,pr_p)):
    ax.text(i-w/2, f+0.005, f'{f:.3f}', ha='center', fontsize=9, fontweight='bold')
    ax.text(i+w/2, p+0.005, f'{p:.3f}', ha='center', fontsize=9, fontweight='bold')
ax.set_xticks(x); ax.set_xticklabels(pnames, fontsize=10, rotation=10)
ax.set_ylim(0,1.1); ax.set_ylabel('Score', fontsize=12)
ax.set_title('Pipeline Order Effect on F1 and PR-AUC (UCI Epileptic)', fontsize=13, fontweight='bold')
ax.legend(['F1','PR-AUC'], fontsize=11)
ax.axvline(1.5, color='gray', ls='--', lw=1)
ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('/content/plots/pipeline_comparison.png', dpi=150)
plt.show()
```
